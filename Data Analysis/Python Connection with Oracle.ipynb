```python
dbdetails = 'hr/atharv2210$@localhost:1521/ORCL'
create_table = "create table tips (total_bill number(10,2), tip number(10,2), sex varchar2(10), smoker varchar2(10), day varchar2(10), time varchar2(10), dsize number)"
selecttable = "select count(*) from tips"
con = cx_Oracle.connect(dbdetails)
cur = con.cursor()
cur.execute(create_table)
rows = [tuple(x) for x in df.values]
cur.executemany("insert into tips (total_bill, tip, sex, smoker, day, time, dsize) values (:1,:2,:3,:4,:5,:6,:7)", rows)
con.commit()
cur.close()
con.close()
```

```python
con = cx_Oracle.connect(dbdetails)
cur = con.cursor()
alter_table = "alter table tips add discount int"
cur.execute(alter_table)
cur.close()
con.close()
```

```python
con = cx_Oracle.connect(dbdetails)
cur = con.cursor()
discount = int(input('Enter the discount: '))
total_bill_range = float(input('Enter the bill amount: '))
query = "update tips set discount =%d where total_bill > %f"
cur.execute(query %(discount, total_bill_range))
con.commit()
cur.close()
con.close()
```
